```python
import torch
from tqdm import tqdm
import transformer_lens
import transformer_lens.utils as utils
from transformer_lens.hook_points import (
    HookPoint,
)  # Hooking utilities
from transformer_lens import HookedTransformer
import matplotlib.pyplot as plt
from transformers import AutoTokenizer

torch.set_grad_enabled(False)
device = utils.get_device()
print(f"Using device {device}")

def imshow(tensor, xaxis="", yaxis="", **kwargs):
    fig, ax = plt.subplots()
    np_tensor = utils.to_numpy(tensor)
    cax = ax.imshow(np_tensor, cmap="RdBu", vmin=np_tensor.min(), vmax=np_tensor.max(), **kwargs)
    ax.set_xlabel(xaxis)
    ax.set_ylable(yaxis)
    fig.colorbar(cax, ax=ax)
    fig.show()
```

```python
model = HookedTransformer.from_pretrained("gpt2-small", device=device)
```

```python
# llama = HookedTransformer.from_pretrained("Llama-2-13b-chat", device=device)
qwen = HookedTransformer.from_pretrained("Qwen/Qwen2.5-7B-Instruct", device=device)
```

```python
tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen2.5-7B-Instruct")
chat = [
    {
        "role": "user",
        "content": "What is the fastest animal?"
    }
]

inputs = tokenizer.apply_chat_template(chat, tokenize=False)
print(inputs)
```

```python
qwen.generate(inputs, max_new_tokens=32)
```
